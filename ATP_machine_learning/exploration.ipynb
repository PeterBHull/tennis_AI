```python
import pandas as pd
import os
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
datadir = '../'
files = os.listdir(datadir)

files = [f for f in files if 'atp_matches_' in f]
df_list = []


for f in files:
    df = pd.read_csv(os.path.join(datadir,f))
    # print(df.head(10))
    df_list.append(df)


data = pd.concat(df_list)
```
